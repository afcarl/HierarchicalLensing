```python
%matplotlib inline
import emcee
import gaussian_likelihood as gl
import plotutils.parameterizations as par
import plotutils.plotutils as pu
import plotutils.runner as pr
from pylab import *
import triangle as tri
```

```python
mu = randn(5)
covm = randn(5,5)
covm = np.dot(covm, covm.T)
```

```python
data = multivariate_normal(mu, covm, size=1000)
```

```python
logl = gl.GaussianLikelihood(data)
p0 = logl.to_params(zeros(logl.nparams))
p0['mean'] = mu
p0['cov_params'] = par.cov_parameters(covm)
p0 = p0.reshape((1,)).view(float)
p0
```

```python
sampler = emcee.EnsembleSampler(128, logl.nparams, logl)
runner = pr.EnsembleSamplerRunner(sampler, p0 + 1e-3*randn(128, logl.nparams))
runner.run_to_neff(16)
```

```python
pu.plot_emcee_chains(runner.chain)
```
