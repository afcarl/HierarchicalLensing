```python
%matplotlib inline
import bz2
import emcee
import plotutils.plotutils as pu
import plotutils.runner as pr
from pylab import *
import scipy.optimize as so
import triangle as tri
import wl_likelihood as wl
```

```python
rs = linspace(0, 10, 100)[1:]
kappas1 = wl.NFW(rs, 1e15, 2, 0.5, 1.2)
kappas2 = wl.NFW(rs, exp(15.5*log(10)), 2, 0.5, 1.2)
plot(rs, kappas1, label=r'$10^{15}\,M_\odot$')
plot(rs, kappas2, label=r'$10^{15.5}\,M_\odot$')
xlabel(r'$r$ (Mpc)')
ylabel(r'$\kappa(r)$')
legend(loc='upper right')
```

## Generate some WL Fake Data and Fit

```python
log_m200 = log(1e15)
c = 10.0
p0 = array([log_m200, c])
zl = 0.5
zs = 1.2
rs = linspace(0, 2, 20)[1:]
shears = wl.NFW(rs, exp(log_m200), c, zl, zs)
dshear = 0.05
errorbar(rs, shears, yerr=dshear)
axis(ymin=0)
```

```python
logpost = wl.WeakLensingLikelihood(rs, shears, dshear, zl, zs)
pbest = so.fmin_powell(lambda x: -logpost(x), p0)
sampler = emcee.EnsembleSampler(32, 2, logpost)
runner = pr.EnsembleSamplerRunner(sampler, pbest + 1e-3*randn(32, 2))
```

```python
runner.run_to_neff(2048/32)
runner.save_state('../runs/wl_test')
```

```python
pu.plot_emcee_chains(runner.chain, mean=False, truths=p0)
tri.corner(runner.thin_flatchain, truths=p0, quantiles=[0.05, 0.95], labels=[r'$\ln M_{200}$', r'$c$'])
```
